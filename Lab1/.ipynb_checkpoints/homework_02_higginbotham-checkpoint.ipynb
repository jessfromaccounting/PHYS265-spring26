### Question 2
---

```python
myList1 = [3, 4.66, "a small puppy dog"]
print(myList1)
print("This is the original list.")
how_many_to_remove = 1
if how_many_to_remove in [1, 2, 3]:
    for i in range(how_many_to_remove):
        myList1.pop();
else:
    print("Invalid number, number must be between 1 and 3.")

print(myList1)
print("This is the final list after removing the specified elements.")
```

```python
import math
print("++++++++++++++++++++++")
print("| n|  pi**n |   e**n |")
print("++++++++++++++++++++++")
for n in range(1,13):
    pi_num = math.pi**n
    e_num = math.e**n
    print(f"|{n:2}|{pi_num:9.3f}|{e_num:8.3f}|")
print("++++++++++++++++++++++")
```

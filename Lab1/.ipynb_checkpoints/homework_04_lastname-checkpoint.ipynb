### Question 1
---

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.exp(x)

def g(x):
    return 1 + x + 0.5*x**2

xx = np.linspace(0,1,75)
yy1 = np.zeros(75)
yy2 = np.zeros(75)

for i in range(len(xx)):
    yy1[i] = f(xx[i])

for i in range(len(xx)):
    yy2[i] = g(xx[i])

fig, ax = plt.subplots(figsize=(5,4))
ax.plot(xx,yy1,label=r'$e^x$',ls='dashed',c='r')
ax.plot(xx,yy2,label=r'$1+x+\frac{1}{2}x^2$',lw=0,marker='o',ms=3,c='black')
ax.set_xlim(-0.2,1.2)
ax.set_ylim(0.75,3)
ax.set_xlabel('x')
ax.set_ylabel('y')
fig.legend(loc='upper left')
fig.tight_layout()
```

### Question 2
---

```python
def myfunction(x):
    return x**3 - 4*x

xx = np.linspace(-2,2,50)
yy1 = np.zeros(50)
yy2 = np.zeros(50)

for i in range(len(xx)):
    yy1[i] = myfunction(xx[i])

for i in range(len(xx)):
    yy2[i] = 2*xx[i]

fig, ax = plt.subplots()
ax.plot(xx,yy1, label=r'$x^3-4x$',c='r',marker='d')
ax.plot(xx,yy2,label=r'$2x$',ls='dashed',c='blue')
ax.set_xlabel('X-axis',size=16)
ax.set_ylabel('Y-axis',size=16)
ax.set_title("Jess' Cool Graph")
fig.legend()
```

### Question 3
---

```python
def f(x,y):
    return 1/np.sqrt((x-3.0)**2 + (y+0.7)**2)

xx = np.linspace(-5.0,5.0,50)
yy = np.zeros(50)

for i in range(len(xx)):
    yy[i] = f(xx[i], -1)

fig, ax = plt.subplots()
ax.plot(xx,yy, label=r'$\frac{1}{\sqrt{(x-3)^2+(y+0.7)^2}}$')
ax.set_xlabel('X-axis',size=16)
ax.set_ylabel('Y-axis',size=16)
fig.legend()
```

### Question 4
---

```python
def piecewise(x):
    y = np.piecewise(x, [x > 0, x <= 0], [x**2, x**3])
    return y

xx = np.linspace(-3,4,70)
yy = np.zeros(70)

for i in range(len(xx)):
    yy[i] = piecewise(xx[i])

fig, ax = plt.subplots()
ax.plot(xx,yy, label=r'$x^2  [x>0],  x^3 [x \leq 0]$')
ax.set_xlabel('X-axis',size=16)
ax.set_ylabel('Y-axis',size=16)
fig.legend()
```
